```python
df=df.dropna()
```

```python
df.rename(columns={"Type 1": "Type_1", "Type 2": "Type_2", "Sp. Atk": "Sp_Atk","Sp. Def":"Sp_Def"},inplace=True)
df
```

```python
mean=df["Total"].mean()
max=df["Total"].max()
min=df["Total"].min()
a=[max,min,mean]
print("max=",max,"\nmin=",min,"\nmean=",mean)
```

```python
dff = pd.DataFrame({
    'X_axis': ['Mean', 'Max_value', 'Min_value'],
    'Values_Mean_Max_Min': [mean, max, min],
})
dff.plot(x="X_axis", y=["Values_Mean_Max_Min"], kind="bar")
```

```python
Sp_Atk=df["Sp_Atk"]
Sp_Def=df["Sp_Def"]

```
