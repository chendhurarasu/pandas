```python
avg_nom_loss=auto['normalized-losses'].astype("float").mean(axis=0)
print("average", avg_nom_loss)
```

```python
auto['normalized-losses'].replace(np.nan,avg_nom_loss,inplace=True)
```

```python
avg_bore=auto['bore'].astype("float").mean(axis=0)
print("avrage",avg_bore)
```

```python
auto['bore'].replace(np.nan,avg_bore,inplace=True)
```

```python
auto['bore'].replace(np.nan,'a',inplace=True)
```

```python
auto.dropna(subset=['price'],axis=0,inplace=True)
auto.reset_index(drop=True,inplace=True)
```

```python
auto.shape
```

```python
auto.isnull().sum()
```

```python

auto['stroke'].replace(np.nan,auto['stroke'].astype('float').mean(axis=0),inplace=True)
auto['horsepower'].replace(np.nan,auto['horsepower'].astype('float').mean(axis=0),inplace=True)
auto['peak-rpm'].replace(np.nan,auto['peak-rpm'].astype('float').mean(axis=0),inplace=True)
```

```python
auto['num-of-doors'].replace(np.nan,'four',inplace=True)
```

```python
auto.isnull().sum()
```

```python
auto[["bore","stroke","price","peak-rpm","horsepower"]]=auto[["bore","stroke","price","peak-rpm","horsepower"]].astype("float")
```

```python
auto[['normalized-losses']]=auto[['normalized-losses']].astype('float')
```

```python
hp_zscore=np.abs(stats.zscore(auto['normalized-losses']))
hp_zscore
```

```python
threshold=hp_zscore.max()
index_val=np.where(hp_zscore>=threshold)
auto.iloc[index_val]['normalized-losses']
```

```python
nd=auto.drop(index=auto.iloc[index_val]['normalized-losses'].index)
```
